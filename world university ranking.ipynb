```python
import pandas as pd

# Load the dataset (adjust the file path as necessary)
df = pd.read_csv('TIMES_WorldUniversityRankings_2024.csv')  # replace with your dataset path

# Display the first few rows and check the data types
print(df.head())
print(df.dtypes)

# Convert relevant columns to numeric, forcing errors to NaN
df['stats_number_students'] = pd.to_numeric(df['stats_number_students'], errors='coerce')
df['stats_student_staff_ratio'] = pd.to_numeric(df['stats_student_staff_ratio'], errors='coerce')

# Now calculate the student-faculty ratio
# Check if 'stats_student_staff_ratio' is zero to avoid division by zero
df['student_faculty_ratio'] = df['stats_number_students'] / df['stats_student_staff_ratio'].replace(0, pd.NA)

# Sort by student-faculty ratio
student_faculty_analysis = df[['name', 'student_faculty_ratio']].sort_values(by='student_faculty_ratio')

# Display universities with the best and worst student-faculty ratios
print("Universities with the best student-faculty ratios:")
print(student_faculty_analysis.head(10))  # Best
print("\nUniversities with the worst student-faculty ratios:")
print(student_faculty_analysis.tail(10))  # Worst
```
